# Data Prep

```python
# source = https://www.kaggle.com/datasets/adamsmith852/life-expectancy-data-gho
!gdown --id 1jNP7r8_ZBT0EBBfXsc6M5w4JEaCO0WxP
FILE = '/content/LifeExpectancy.csv' #2019, male n female dipisah
```

```python
# import file utuh dari 2 tahun
import pandas as pd
df = pd.read_csv(FILE)
df
```

```python
# ambil yang hanya 2019
grouped = df.groupby(['Year'])

df = grouped.get_group(2019)
df
```

```python
# cek kolom yang 0 semua
zero_cols = [i for i in df.columns if df[i].max() == df[i].min() == 0]
print(zero_cols)

#cek kolom yang null semua
null_cols = [i for i in df.columns if df[i].isna().sum()==len(df)]
print(null_cols)
```

```python
# drop kolom yang kosong atau 0 smua, sama tahun
df.drop(columns=[i for i in null_cols], inplace=True)
df.drop(columns=[i for i in zero_cols], inplace=True)
df
```

```python
#print kolom yg masih ada nilai kosong
nan_cols = [i for i in df.columns if df[i].isnull().any()]
nan_cols
```

```python
df[nan_cols].describe()
```

```python
# #interpolate, isi missing value
# df['Life Expectancy at birth'] = df['Life Expectancy at birth'].interpolate()
# df['Alcohol'] = df['Alcohol'].interpolate()

# Handling missing value dengan diisi rata2 (mean imputation)
median_value = df['Life Expectancy at birth'].median()
df['Life Expectancy at birth'].fillna(median_value, inplace=True)

#Alcohol
mean_value = df['Alcohol'].mean()
df['Alcohol'].fillna(mean_value, inplace=True)
```

```python
#pisahin male n female ke df beda, ambil male doang
grouped = df.groupby(['Gender'])
df = grouped.get_group("Male")
```

```python
# simpan dulu labelnya biar bisa diplot nnti
countries = [i for i in df['Country']]
countries
```

```python
# buang label
df = df.drop([i for i in df.columns if i in ['Country', 'Year', 'Gender']], axis=1)
df
```

```python
#normalisasi data, ubah dari 0 ke 1 nilainya
import numpy as np
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df = scaler.fit_transform(df)
```

```python
# PCA, dimensi reduksi biar bisa di plot
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

principal_components = pca.fit_transform(df)

ex_var_ratio = pca.explained_variance_ratio_
print(ex_var_ratio)

component_1 = principal_components[:, 0]
component_2 = principal_components[:, 1]
```

```python
df = pd.DataFrame({'Feature1': component_1, 'Feature2': component_2})
df
```

# DBScan

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import DBSCAN

silh = []
for c in np.arange(0.1, 1, 0.1):
  # apply DBSCAN clustering
  dbscan = DBSCAN(eps=c, min_samples=5)
  dbscan.fit(df)

  # retrieve the labels and number of clusters
  labels = dbscan.labels_
  n_clusters = len(set(labels)) - (1 if -1 in labels else 0) # exclude noise points (-1 label)

  sil_avg = silhouette_score(df, labels)
  silh.append(sil_avg)

  print('Nilai e: ', c)
  print('Silhouette Average Score: ', sil_avg, '\n')

print(max(silh), 2+silh.index(max(silh)))
```

```python
import matplotlib.pyplot as plt

plt.title('DBScan Silhouette')
plt.xlabel("e")
plt.ylabel("Silhouette Coefficient")
plt.xticks(np.arange(0.1, 1, 0.1))
plt.plot(np.arange(0.1, 1, 0.1), silh, 'b.-')
```

```python
from sklearn.cluster import DBSCAN

# set the parameters for DBSCAN
epsilon = 0.9 # the maximum distance between two points to be considered in the same neighborhood
min_samples = 4 # the minimum number of points in a neighborhood to form a cluster

# apply DBSCAN clustering
dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)
dbscan.fit(df)

# retrieve the labels and number of clusters
labels = dbscan.labels_
n_clusters = len(set(labels))

# n_clusters = len(set(labels)) - (1 if -1 in labels else 0) # exclude noise points (-1 label)

# print the results
print("Number of clusters:", n_clusters)
print("Cluster labels:", labels)
```

```python
df_result = pd.DataFrame({'Negara': countries, 'Cluster': labels})
df_result[df_result['Cluster']==-1]
```

```python
df_result.to_csv('DB.csv', index=False)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Feature1'], df['Feature2'], c=labels)
plt.title("DBSCAN Clustering Male Results")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

```python
from sklearn.metrics import calinski_harabasz_score
# compute the Calinski-Harabasz score to evaluate the clustering performance
ch_score = calinski_harabasz_score(df, labels)

# print the Calinski-Harabasz score
print("Calinski-Harabasz score: ", ch_score)
```

```python
from sklearn.metrics import davies_bouldin_score #tpi ga cocok buat dbscan

dbi = davies_bouldin_score(df, labels)
print("Davies-Bouldin Index:", dbi)
```

```python
from sklearn.metrics import silhouette_score
silhouette = silhouette_score(df, labels)
print("Silhouette coefficient:", silhouette)
```

# K-Means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans_kwargs = { #edit parameter
    "init": "k-means++",
    "max_iter": 300,
    "random_state": 42,
}

wcss = [] #nyari wcss tiap iter
silh = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k, n_init=10, **kmeans_kwargs)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

    sil_avg = silhouette_score(df, kmeans.labels_)
    silh.append(sil_avg)

    print('\nNilai k: ', k)
    print('Nilai WCSS: ', kmeans.inertia_)
    print('Silhouette Average Score: ', sil_avg, '\n')
```

```python
import matplotlib.pyplot as plt
plt.title('K-Means WCSS')
plt.plot(range(2, 20), wcss, 'b.-')
plt.xticks(range(2, 20))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
index = []
for i in range(2, 20):
  index.append(i)

plt.title('K-Means Silhouette')
plt.xlabel("K")
plt.ylabel("Silhouette Coefficient")
plt.xticks(range(2,20))
plt.plot(range(2,20), silh, 'b.-')
```

```python
from sklearn.cluster import KMeans

# set the number of clusters for K-means
num_clusters = 6

# apply K-means clustering
kmeans = KMeans(n_clusters=num_clusters)
labels = kmeans.fit_predict(df)

# print the cluster labels
# print('Number of clusters: ', len(np.unique(labels)))
print('Number of clusters: ', num_clusters)
print('Cluster labels: ', labels)
```

```python
df_result = pd.DataFrame({'Negara': countries, 'Cluster': labels})
df_result[df_result['Cluster']==4]
```
